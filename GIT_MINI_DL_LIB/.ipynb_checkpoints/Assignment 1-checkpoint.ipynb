# Notes:
## Numeric Instability of Softmax: (see layers.py/softmax_loss())
$$\begin{equation}
    \frac{e^{f_{y_i}}}{\sum_{j}e^{f_j}} = \frac{Ce^{f_{y_i}}}{C\sum_{j}e^{f_j}} = \frac{e^{f_{y_i} + \log C}}{\sum_{j}e^{f_j + \log C}}
\end{equation}$$ where commonly $\log C = -\max_j f_j$, it shifts the value to make highest value be zero

```python
class FullyConnectedNet(object):
    """A fully connected neural network with arbitrary number of hidden layers,
    ReLU nonlineaities and a softmax loss function. Optional layers include dropout
    and batch normalization. 
    
    Network architecture of L layers:
        {affine - [batch norm] - relu - [dropout]} x (L - 1) - affine - softmax
        where {...} repeated (L - 1) times.
        
    The trainable parameters are stored in dictionary self.params that maps
    parameter names to numpy arrays.
    """
    
    def __init__(self, 
                 hidden_dims, 
                 input_dim=1*28*28, 
                 num_classes=10,
                 dropout=0,
                 use_batchnorm=False,
                 reg=0.0,
                 weight_scale=1e-2,
                 dtype=np.float32,
                 seed=None):
        """Initialize a new FullyConnectedNet.
        
        Args:
            hidden_dims: A list of integers, size of each hidden layer
            input_dim: An integer, input dimension
            num_classes: An integer, number of classes
            dropout: A float in [0, 1], dropout strength. If dropout=0: disable dropout
            use_batchnorm: A bool, True: use batch normalization
            reg: A float, L2 regularization strength
            weight_scale: A float, standard deviation for random weights initialization
            dtype: A numpy datatype, all computations use this datatype, e.g. np.float32
            seed: If not None: pass it to dropout to be deterministic, for gradient check
        """
        
        self.use_batchnorm = use_batchnorm
        self.use_dropout = dropout > 0
        self.reg = reg
        self.num_layers = 1 + len(hidden_dims) # plus final affine layer before softmax
        self.dtype = dtype
        self.params = {}
        
        # Initialize network parameters:
        # W: normal distribution with standard deviation = weight_scale
        # b: zero
        # Batch norm: gamma(scale): ones, beta(shift):zeros
        pre_dim = input_dim
        for layer_num, hidden_dim in enumerate(hidden_dims, start=1):
            # Initialize W and b
            self.params['W{}'.format(layer_num)] = np.random.normal(0, weight_scale, [pre_dim, hidden_dim])
            self.params['b{}'.format(layer_num)] = np.zeros(hidden_dim)
            # Add batch normalization when flag is True
            if self.use_batchnorm and layer_num < self.num_layers:
                self.params['gamma{}'.format(layer_num)] = np.ones(hidden_dim)
                self.params['beta{}'.format(layer_num)] = np.zeros(hidden_dim)
            pre_dim = hidden_dim
        
        # Final affine layer before softmax
        self.params['W{}'.format(self.num_layers)] = np.random.normal(0, 
                                                                      weight_scale, 
                                                                      [hidden_dims[-1], num_classes])
        self.params['b{}'.format(self.num_layers)] = np.zeros(num_classes)
        
        # Dropout param should be passed to each of its layer for probability and mode (train/test)
        self.dropout_param = {}
        if self.use_dropout:
            self.dropout_param = {'mode': 'train', 'p': dropout}
            if seed is not None:
                self.dropout_param['seed'] = seed
                
        # Batch norm param should be passed to each of its layer
        # To keep track of running means and variances
        self.bn_params = []
        if self.use_batchnorm:
            self.bn_params = [{'mode': 'train'} for i in range(self.num_layers - 1)]
            
        
        # Cast all parameters to the correct datatype
        for key, value in self.params.items():
            self.params[key] = value.astype(dtype)
            
    def loss(self, X, y=None):
        """Compute loss and gradient for a minibatch of data
        
        Args:
            X: A numpy array, shape [N, d_1, ..., d_k], input data
            y: A numpy array, shape [N, ], y[i] gives the label for X[i]
        
        Returns:
            If y is None: Test time forward pass
                scores: A numpy array, shape [N, C]
                
            If y is not None: Training time forward and backward pass
                loss: A float
                grads: A dictionary, same keys as self.params, mapping parameter names
                        to gradients of loss w.r.t parameter
        """
        
        # Ensure correct datatype
        X = X.astype(self.dtype)
        
        mode = 'test' if y is None else 'train'
        
        # Set dropout and batch norm params for train/test mode
        if self.dropout_param is not None:
            self.dropout_param['mode'] = mode
        if self.use_batchnorm:
            for bn_param in self.bn_params:
                bn_param['mode'] = mode
                
        # Forward pass for fully connected net
        data_in = X
        caches = {} ##########Is it really {} not [] ???????????#######################????????????????/
        for i in range(self.num_layers):
            W_i = self.params['W{}'.format(i + 1)]
            b_i = self.params['b{}'.format(i + 1)]
            
            if i + 1 != self.num_layers: # Before final affine layer
                if self.use_batchnorm:
                    gamma_i = self.params['gamma{}'.format(i + 1)]
                    beta_i = self.params['beta{}'.format(i + 1)]
                    bn_params_i = self.bn_params[i]
                    # Forward pass
                    data_in, caches[i] = affine_batchnorm_relu_forward(data_in, 
                                                                       W_i, 
                                                                       b_i, 
                                                                       gamma_i, 
                                                                       beta_i, 
                                                                       bn_params_i)
                else:
                    data_in, caches[i] = affine_relu_forward(data_in, W_i, b_i)

                if self.use_dropout:
                    data_in, dropout_cache = dropout_forward(data_in, self.dropout_param)
                    caches[i] = caches[i] + (dropout_cache, )
            else: # Final affine layer before softmax
                scores, caches[i] = affine_forward(data_in, W_i, b_i)
                
        # If test mode, return early
        if mode == 'test':
            return scores
        
        # Backward pass for fully connected net
        # Batch norm: No need to regularize the scale and shift parameters
        grads = {}
        loss, dout = softmax_loss(scores, y)

        for i in range(self.num_layers)[::-1]: # backward loop
            if i + 1 == self.num_layers: # Backprop on final affine layer before softmax
                dout, grads['W{}'.format(i + 1)], grads['b{}'.format(i + 1)] = affine_backward(dout, caches[i])
            else:
                tmp_cache = caches[i]
                
                if self.use_dropout: # independent test if use dropout
                    dout = dropout_backward(dout, tmp_cache[-1])
                    tmp_cache = tmp_cache[:-1]
                if self.use_batchnorm:
                    result = affine_batchnorm_relu_backward(dout, tmp_cache)
                    dout, grads['W{}'.format(i + 1)], grads['b{}'.format(i + 1)] = result[:3]
                    grads['gamma{}'.format(i + 1)], grads['beta{}'.format(i + 1)] = result[3:]
                else:
                    result = affine_relu_backward(dout, tmp_cache)
                    dout, grads['W{}'.format(i + 1)], grads['b{}'.format(i + 1)] = result
                    
            # Regularization
            loss += 0.5*self.reg*np.sum(self.params['W{}'.format(i + 1)]**2)
            grads['W{}'.format(i + 1)] += self.reg*self.params['W{}'.format(i + 1)]
                    
        return loss, grads
```

## MLP

```python
N, D, H, C = 3, 5, 50, 7
X = np.random.randn(N, D)
y = np.random.randint(C, size=N)

std = 1e-2
model = FullyConnectedNet(hidden_dims=[H], 
                          input_dim=D, 
                          num_classes=C, 
                          weight_scale=std)

print('Testing initialization ...')
W1_std = abs(model.params['W1'].std() - std)
b1 = model.params['b1']
W2_std = abs(model.params['W2'].std() - std)
b2 = model.params['b2']
assert W1_std < std/10, 'First layer weights do not seem right'
assert np.all(b1 == 0), 'First layer biase do not seem right'
assert W2_std < std/10, 'Second layer weights do not seem right'
assert np.all(b2 == 0), 'Second layer biase do not seem right'

print('Testing test-time forward pass ...')
model.params['W1'] = np.linspace(-0.7, 0.3, num=D*H).reshape(D, H)
model.params['b1'] = np.linspace(-0.1, 0.9, num=H)
model.params['W2'] = np.linspace(-0.3, 0.4, num=H*C).reshape(H, C)
model.params['b2'] = np.linspace(-0.9, 0.1, num=C)
X = np.linspace(-5.5, 4.5, num=N*D).reshape(D, N).T
scores = model.loss(X)
correct_scores = np.asarray(
  [[11.53165108,  12.2917344,   13.05181771,  13.81190102,  14.57198434, 15.33206765,  16.09215096],
   [12.05769098,  12.74614105,  13.43459113,  14.1230412,   14.81149128, 15.49994135,  16.18839143],
   [12.58373087,  13.20054771,  13.81736455,  14.43418138,  15.05099822, 15.66781506,  16.2846319 ]])
scores_diff = np.abs(scores - correct_scores).sum()

assert scores_diff < 1e-5, 'Problem with test-time forward pass'

print('Testing training loss (no regularization)')
y = np.array([0, 5, 1])
loss, grads = model.loss(X, y)
correct_loss = 3.4702243556
assert abs(loss - correct_loss) < 1e-7, 'Problem with training-time loss'

model.reg = 1.0
loss, grads = model.loss(X, y)
correct_loss = 26.5948426952
assert abs(loss - correct_loss) < 1e-7, 'Problem with regularization loss'

for reg in [0.0, 0.7]:
    print('Running numeric gradient check with reg = ', reg)
    model.reg = reg
    loss, grads = model.loss(X, y)
    
    for name in sorted(grads):
        f = lambda _: model.loss(X, y)[0]
        grad_num = eval_numerical_gradient(f, model.params[name])
        print('{} relative error: {:.2e}'.format(name, rel_error(grad_num, grads[name])))
```

## Compare SGD and SGDMomentum
